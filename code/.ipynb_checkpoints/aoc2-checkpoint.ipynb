```python
his_points = {"A": 1, "B": 2, "C": 3}
my_points = {"X": 1, "Y": 2, "Z": 3}
with open(r"C:\Users\TomCortenRiskQuest\Documents\haven_question\code\aoc_data\input.txt",'r') as f:
    lines = [line.rstrip() for line in f]   
```

```python
score = 0
for play in lines:
    his, my = play.split(" ")
    if my_points[my]==his_points[his]:
        score += my_points[my] +3
    if my_points[my]-his_points[his]==1:
        score += my_points[my] + 6
        print(his, my, my_points[my])
    if (my =='X' and his == "C"):
        score += my_points[my] + 6
    if his_points[his]-my_points[my]==1:
        score += my_points[my] 
    if (my =='Z' and his == "A"):
        score += my_points[my] 
```

```python
score = 0
for play in lines:
    his, my = play.split(" ")
    if my == "Y":
        score += his_points[his] +3
    if (my == "X" and his != "A"):
        score += his_points[his] - 1
    if (my == "X" and his == "A"):
        score += his_points[his] + 2 
    if (my == "Z" and his != "C"):
        score += his_points[his] + 1 + 6
    if (my == "Z" and his == "C"):
        score += his_points[his] + 6 - 2 
```
